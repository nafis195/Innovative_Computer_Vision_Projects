**Class 2**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/content/alzheimer-disease-T1.jpg')
```

```python
plt.imshow(image)
```

```python
plt.show()
```

```python
brightness = np.mean(image)
print(f"Brightness (Average pixel intensity): {brightness}")
```

```python
contrast = np.std(image)
print(f"Contrast (Standard deviation of pixel intensities): {contrast}")
```

```python
print('Shape of the image:', image.shape)
```

```python
print("Matrix of the image first 5x5 pixels:")
print(image[:5, :5])
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
gray_image.shape
```

```python
alpha = 2
beta = 30
adjusted_image = cv2.convertScaleAbs(gray_image, alpha=alpha, beta=beta)
```

```python
# plot the original and adjusted images
plt.figure(figsize=(10, 5))

#plot original image
plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Original Image')
plt.axis('Off')

# plot adjusted image
plt.subplot(1, 2, 2)
plt.imshow(adjusted_image, cmap='gray')
plt.title('Adjusted Image')
plt.axis('Off')

plt.show()
```

```python
# calculate the histogram of the image
histogram = cv2.calcHist([image], [0], None, [256], [0, 256])

# plot the histogram
plt.plot(histogram)
plt.title('Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```

```python
# apply histogram equalization
equalized_image = cv2.equalizeHist(gray_image)

# calculate the histogram of the equalized image
equalized_histogram = cv2.calcHist([equalized_image], [0], None, [256], [0, 256])

# plot the images and histogram in a 2x2 grid
plt.figure(figsize=(12, 12))

# original image
plt.subplot(2, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Original Image')
plt.axis('Off')

# equalized image
plt.subplot(2, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.axis('Off')

# original histogram
plt.subplot(2, 2, 3)
plt.plot(histogram)
plt.title('Original Image Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

# equalized histogram
plt.subplot(2, 2, 4)
plt.plot(equalized_histogram)
plt.title('Equalized Image Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

# display the plots
plt.tight_layout()
plt.show()
```

```python
gamma = 3.2
gamma_corrected_image = np.array(255 * (gray_image / 255) ** gamma, dtype='uint8')
```

```python
plt.figure(figsize = (10, 5))

plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Original Image')
plt.axis('Off')

plt.subplot(1, 2, 2)
plt.imshow(gamma_corrected_image, cmap='gray')
plt.title('Gamma Corrected Image')
plt.axis('Off')

plt.show()
```

**CLASS 3**

```python
image = cv2.imread('/content/class3_blurry.jpeg')
```

```python
plt.imshow(image)
plt.show()
```

```python
print("shape: ", image.shape)
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
average_filtered = cv2.blur(image, (5, 5))
average_filtered_rgb = cv2.cvtColor(average_filtered, cv2.COLOR_BGR2RGB)
```

```python
# plot original and average filtered images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('Off')

plt.subplot(1, 2, 2)
plt.imshow(average_filtered_rgb)
plt.title('Average (Mean) Filtered Image')
plt.axis('Off')

plt.show()
```

```python
gaussian_filtered = cv2.GaussianBlur(image, (5, 5), 0)
# gaussian_filtered_rgb = cv2.cvtColor(gaussian_filtered, cv2.COLOR_BGR2RGB)
```

```python
# plot original and gaussian filtered images
plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('Off')

plt.subplot(1, 2, 2)
plt.imshow(gaussian_filtered)
plt.title('Gaussian Filtered Image')
plt.axis('Off')

plt.show()
```

```python
median_filtered = cv2.medianBlur(image_rgb, 5)
```

```python
# plot original and gaussian filtered images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('Off')

plt.subplot(1, 2, 2)
plt.imshow(median_filtered)
plt.title('Median Filtered Image')
plt.axis('Off')

plt.show()
```

```python
bilateral_filtered = cv2.bilateralFilter(image_rgb, d=9, sigmaColor=75, sigmaSpace=75)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('Off')

plt.subplot(1, 2, 2)
plt.imshow(bilateral_filtered)
plt.title('Bilateral Filtered Image')
plt.axis('Off')

plt.show()
```

```python
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

sharpened_image = cv2.filter2D(src = image, ddepth = -1, kernel = kernel)
sharpened_image_rgb = cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB)
```

```python
# plot original and sharpened images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('Off')

plt.subplot(1, 2, 2)
plt.imshow(sharpened_image_rgb)
plt.title('Sharpened Image')
plt.axis('Off')

plt.show()
```
